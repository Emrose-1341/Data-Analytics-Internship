```python
import pandas as pd
```

```python
df = pd.read_excel('../data/IL YOUTH DATASET FROM FY23-25.xlsx', sheet_name='Dataset', 
                   dtype={'PERSON ID': str, 'REMOVAL ZIP CODE': str, 'PLACEMENT ZIP CODE': str})
```

```python
print(df.shape) 
df.head()
```

```python
print(f"Total records: {len(df['PERSON ID'])}")
print(f"Unique youth: {df['PERSON ID'].nunique()}")
```

```python
print("\nCounts by age:")
age_counts = df.groupby('CURRENT AGE')['PERSON ID'].agg(['count', 'nunique'])
age_counts.columns = ['Total records', 'Unique youth']
age_counts['Records per youth'] = age_counts['Total records'] / age_counts['Unique youth']
print(age_counts)
```

```python
print("\nCounts by gender:")
gender_counts = df.groupby('GENDER')['PERSON ID'].agg(['count', 'nunique'])
gender_counts.columns = ['Total records', 'Unique youth']
print(gender_counts)
```

```python
print(f"\nTotal unique counties: {df['COMMITMENT COUNTY'].nunique()}")
print("\nCounts by county:")
county_counts = df.groupby('COMMITMENT COUNTY')['PERSON ID'].agg(['count', 'nunique'])
county_counts.columns = ['Total records', 'Unique youth'] 
print(county_counts)
```

```python
print(f"\nTotal unique removal zip codes: {df['REMOVAL ZIP CODE'].nunique()}")
print("\nCounts by removal zip code:")
zip_counts = df.groupby('REMOVAL ZIP CODE')['PERSON ID'].agg(['count', 'nunique'])
zip_counts.columns = ['Total records', 'Unique youth']
print(zip_counts)
```

```python
print(f"\nTotal unique placement zip codes: {df['PLACEMENT ZIP CODE'].nunique()}")
print("\nCounts by placement zip code:")
zip_counts = df.groupby('PLACEMENT ZIP CODE')['PERSON ID'].agg(['count', 'nunique'])
zip_counts.columns = ['Total records', 'Unique youth']
print(zip_counts)
```

```python
print(f"\nTotal unique assessment indicators: {df['ASSESSMENT INDICATOR'].nunique()}")
print("\nCounts by assessment indicator:")
assessment_counts = df.groupby('ASSESSMENT INDICATOR')['PERSON ID'].agg(['count', 'nunique'])
assessment_counts.columns = ['Total records', 'Unique youth']
assessment_counts
```

```python
# Count unique values per person for key fields
person_counts = df.groupby('PERSON ID').agg({
    'CURRENT AGE': 'nunique',
    'COMMITMENT COUNTY': 'nunique', 
    'RESPONSIBLE COUNTY': 'nunique',
    'GENDER': 'nunique',
    'REMOVAL ZIP CODE': 'nunique',
    'PLACEMENT ZIP CODE': 'nunique',
    'LOCATION BEGIN DATE': 'nunique',
    'LOCATION END DATE': 'nunique',
    'ASSESSMENT TYPE': 'nunique'
}).reset_index() 

print(person_counts.shape)
```

```python
person_counts.head(10)
```

```python
col = 'ASSESSMENT TYPE'
person_counts[['PERSON ID',col]].sort_values(by=col, ascending=False)#.iloc[0:25,:]
```

```python
### For each column in person_counts, except PERSON ID, display value_counts()
for col in person_counts.columns:
    if col != 'PERSON ID':
        print(f"\nValue counts for {col}:")
        print(person_counts[col].value_counts())
```

```python
### Create a Youth Table that contains all a single row per youth and displays the following fields: 
# Current Age, Commitment County, Responsible County, Gender, Removal Zip Code, Placement Zip Code, Location Begin Date, Location End Date
# Assessment Type will be pivoted with an indicator if it is present for that youth. 

df_youth = df[['PERSON ID', 'CURRENT AGE', 'COMMITMENT COUNTY', 'RESPONSIBLE COUNTY', 'GENDER', 'REMOVAL ZIP CODE', 'PLACEMENT ZIP CODE', 'LOCATION BEGIN DATE', 'LOCATION END DATE']].copy().drop_duplicates().reset_index(drop=True)
print(df_youth.shape)
df_youth.head()
```

```python
df_CANS = df[df['ASSESSMENT TYPE'] == 'CANS'][['PERSON ID']].drop_duplicates().reset_index(drop=True)
df_CANS['CANS'] = 1 
print(df_CANS.shape) 

df_LIFESKILL = df[df['ASSESSMENT TYPE'] == 'LIFESKILL'][['PERSON ID']].drop_duplicates().reset_index(drop=True)
df_LIFESKILL['LIFESKILL'] = 1 
print(df_LIFESKILL.shape) 
```

```python
### Merge with Assessment Types
df_youth = df_youth.merge(df_CANS, on='PERSON ID', how='left') 
df_youth = df_youth.merge(df_LIFESKILL, on='PERSON ID', how='left') 

# Fill CANS and LIFESKILL with 0 where they are NaN
df_youth['CANS'] = df_youth['CANS'].fillna(0)
df_youth['LIFESKILL'] = df_youth['LIFESKILL'].fillna(0)


print(df_youth.shape)
df_youth.head()
```

```python
df_youth[['CANS','LIFESKILL']].value_counts()
```

```python
### Save df_youth to csv in 'bdaic created tables' folder 
df_youth.to_csv('../data/bdaic created tables/df_youth.csv', index=False)
```
