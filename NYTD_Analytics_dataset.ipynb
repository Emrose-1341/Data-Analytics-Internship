```python
import pandas as pd
import boto3
import os
import io

# Initialize S3 client
s3 = boto3.client('s3')

# List all files in the cleaned directory
response = s3.list_objects_v2(Bucket='bdc-public-curated', Prefix='ndacan/nytd/outcomes/cleaned')

# Get the list of files, excluding archive folder
files = [obj['Key'] for obj in response['Contents']  
         if obj['Key'].endswith('.csv') and 'archive' not in obj['Key']]
```

```python
files
```

```python
# Create dictionary to store dataframes
dfs = {}

# Import each CSV file from S3 into dictionary
for file in files:
    if file.endswith('.csv'):
        # Get filename without path and extension to use as dict key
        filename = os.path.basename(file).replace('_Cleaned.csv','')
        
        # Read CSV from S3 using boto3
        obj = s3.get_object(Bucket='bdc-public-curated', Key=file)
        df = pd.read_csv(obj['Body'])
        
        # Store in dictionary
        dfs[filename] = df
```

```python
# Get list of columns for each dataset
column_sets = {name: set(df.columns) for name, df in dfs.items()}

# Find common columns across all datasets
common_columns = set.intersection(*column_sets.values())

# Find columns that appear in some but not all datasets
all_columns = set.union(*column_sets.values())
partial_columns = all_columns - common_columns

print("Columns present in ALL datasets:")
print("\n".join(sorted(common_columns)))
print("\nColumns present in SOME datasets:")
for col in sorted(partial_columns):
    present_in = [name for name, cols in column_sets.items() if col in cols]
    print(f"{col}: Present in {', '.join(present_in)}")
```

```python
# Create Cohort column for each dataset
for name, df in dfs.items():
    # Find the cohort column that matches the pattern FYXXCohort
    cohort_col = [col for col in df.columns if col.startswith('FY') and col.endswith('Cohort')][0]
    # Extract first 4 characters to get FYxx
    df['Cohort'] = cohort_col[:4]
```

```python
# Generate a "Race" column for the 202 and 228 datasets ('Race' is already in the 266 and 297 datasets) 
# Use the following mapping: 
# 'White' = 1 then '1'
# 'BlkAfrAm' = 1 then '2' 
# 'AmIAKN' = 1 then '3'
# 'Asian' = 1 then '4'
# 'HawaiiPI' = 1 then '5'
#sum('White','BlkAfrAm','AmIAKN','Asian','HawaiiPI') > 1 then '6'
#'RaceUnkn' OR 'RaceDcln' = 1 then '99'

for dataset in ['NYTD_Dataset_202', 'NYTD_Dataset_228']:
    df = dfs[dataset]
    
    # Initialize Race column as 0
    df['Race'] = 0
    
    # Apply the race mapping rules
    df.loc[df['White'] == 1, 'Race'] = 1
    df.loc[df['BlkAfrAm'] == 1, 'Race'] = 2
    df.loc[df['AmIAKN'] == 1, 'Race'] = 3
    df.loc[df['Asian'] == 1, 'Race'] = 4
    df.loc[df['HawaiiPI'] == 1, 'Race'] = 5
    
    # Multiple races (sum > 1)
    race_cols = ['White', 'BlkAfrAm', 'AmIAKN', 'Asian', 'HawaiiPI']
    df.loc[df[race_cols].sum(axis=1) > 1, 'Race'] = 6
    
    # Unknown or declined
    df.loc[(df['RaceUnkn'] == 1) | (df['RaceDcln'] == 1), 'Race'] = 99
    
    # Convert to integer
    df['Race'] = df['Race'].astype(float)
```

## For TN Wave 1, 2, and 3 dataframes, update headers and drop columns


```python
# Add wave number suffix to each of the non-identifier headers in wave1_df, wave2_df, and wave3_df in the form of "_w#"
reference_cols = ['OutcmRpt', 'OutcmDte', 'OutcmFCS', 'Race', 'Responded', 'Elig19', 'Elig21']   
pred_cols = ['EmplySklls', 'SocSecrty', 'EducAid', 'PubFinAs', 'PubFoodAs', 'PubHousAs', 'OthrFinAs', 'CnctAdult', 'HighEdCert',
            'Children', 'Marriage', 'Medicaid', 'OthrHlthIn', 'MedicalIn', 'MentlHlthIn', 'PrescripIn']
outcome_cols = ['HighEdCert', 'CurrenRoll', 'CurrFTE', 'CurrPTE', 'Homeless', 'SubAbuse', 'Incarc']


# Add wave number suffix to columns in wave1_df
for col in reference_cols +pred_cols + outcome_cols:
    if col in wave1_df.columns:
        wave1_df = wave1_df.rename(columns={col: f'{col}_w1'})

# Add wave number suffix to columns in wave2_df  
for col in reference_cols +pred_cols + outcome_cols:
    if col in wave2_df.columns:
        wave2_df = wave2_df.rename(columns={col: f'{col}_w2'})

# Add wave number suffix to columns in wave3_df
for col in reference_cols + pred_cols + outcome_cols:
    if col in wave3_df.columns:
        wave3_df = wave3_df.rename(columns={col: f'{col}_w3'})

```
